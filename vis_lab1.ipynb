```python
somedata = pd.DataFrame(data = ser.index.tolist()[1:], columns=ser.index.tolist()[0])
```

```python
data = somedata[["Year", "J-D"]].copy()
```

```python
data =data.astype({"Year":int, "J-D":float})
```

```python
min_temp = data["J-D"].min()
```

```python
mean_tem = data.loc[data["Year"].isin(range(1880,1899))]["J-D"].mean()
mean_tem
```

```python
yr_range = [data["Year"].min(), data["Year"].max()]
temp_range = [data["J-D"].min() - abs(data["J-D"].min()*0.15),
              data["J-D"].max() + abs(data["J-D"].max()*0.15)]

base = alt.Chart(data).mark_circle(color='#f26c08', 
                                   stroke='black',
                                   strokeWidth=1,
                                   size=50
).encode(
    alt.X("Year:Q", scale=alt.Scale(domain=yr_range),axis=alt.Axis( title="", tickCount=12, labelOpacity=0.7,gridOpacity=0.7)), 
    alt.Y("J-D:Q", scale=alt.Scale(domain=temp_range),axis=alt.Axis( title="", tickCount=7, labelOpacity=0.7,gridOpacity=0.7)) 
)

text = (
    alt.Chart(data.loc[data["Year"].isin([1944,2014,2015,2016,2017])])
    .mark_text(dy=0,dx=20, color="black")
.encode(
    alt.X("Year:Q", scale=alt.Scale(domain=yr_range)),
    y="J-D:Q", text=alt.Text("Year:Q"))
)

text_left = (
    alt.Chart(data.loc[data["Year"].isin([1904,1998])])
    .mark_text(dy=0,dx=-20, color="black")
.encode(
    alt.X("Year:Q", scale=alt.Scale(domain=yr_range)),
    y="J-D:Q", text=alt.Text("Year:Q"))
)
big_text = alt.Chart(
    pd.DataFrame({'y': [mean_tem], 'x':[2000]})
).mark_text(
    text="HOTTER THAN THE\n 1880-1899 AVERAGE", 
    color='grey', fontSize=14, lineBreak='\n', 
    align='left', baseline='middle', dy=-30
).encode(x="x:Q", y="y:Q")
big_text2 = alt.Chart(
    pd.DataFrame({'y': [mean_tem], 'x':[2000]})
).mark_text(
    text="COLDER", 
    color='grey', fontSize=14, lineBreak='\n', 
    align='left', baseline='middle', dy=10
).encode(x="x:Q", y="y:Q")

line = alt.Chart(pd.DataFrame({'y': [mean_tem]})).mark_rule(color='grey').encode(y='y')
```

```python
res = (base+text+text_left+line+big_text+big_text2
      ).configure_axis(
labelColor="grey"
).properties(width = 957, height=604, padding={"left": 15, "top": 10, "right": 10, "bottom": 40})
```
